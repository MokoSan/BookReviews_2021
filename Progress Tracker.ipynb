```python
import matplotlib.pyplot as plt
import calendar
```

```python
%matplotlib inline
```

```python
months = [calendar.month_name[i] for i in range(1, 13)]
data   = [4, 4, 6, 3, 0, 0, 0, 0, 0, 0, 0, 0]

plt.bar(months, data)
plt.title(f'Books Read So Far: {sum(data)}; Goal: 45')
plt.xticks(rotation='vertical')
plt.xlabel('Month')
plt.ylabel('# of Books')
plt.show()
```
